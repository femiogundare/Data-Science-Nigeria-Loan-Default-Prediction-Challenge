```python
def age_group(age):
    if (age>=20) and (age<=25):
        return '20-25'
    elif (age>25) and (age<=30):
        return '26-30'
    elif (age>30) and (age<=35):
        return '31-35'
    elif (age>35) and (age<=40):
        return '36-40'
    elif (age>40) and (age<=45):
        return '41-45'
    elif (age>45) and (age<=50):
        return '46-50'
    elif (age>50) and (age<=55):
        return '51-55'
    elif (age>55) and (age<=60):
        return '56-60'
    
train_dem['Age_Bracket'] = train_dem['Age_in_2018'].map(age_group)
test_dem['Age_Bracket'] = test_dem['Age_in_2018'].map(age_group)
```

```python
def x_coord(df):
    long = df[0]
    lat = df[1]
    
    return np.cos(lat)*np.cos(long)

def y_coord(df):
    long = df[0]
    lat = df[1]
    
    return np.cos(lat)*np.sin(long)

def z_coord(df):
    long = df[0]
    lat = df[1]
    
    return np.sin(lat)


train_dem['X-Coordinate'] = train_dem[['longitude_gps', 'latitude_gps']].apply(x_coord, axis=1)
test_dem['X-Coordinate'] = test_dem[['longitude_gps', 'latitude_gps']].apply(x_coord, axis=1)

train_dem['Y-Coordinate'] = train_dem[['longitude_gps', 'latitude_gps']].apply(y_coord, axis=1)
test_dem['Y-Coordinate'] = test_dem[['longitude_gps', 'latitude_gps']].apply(y_coord, axis=1)

train_dem['Z-Coordinate'] = train_dem[['longitude_gps', 'latitude_gps']].apply(z_coord, axis=1)
test_dem['Z-Coordinate'] = test_dem[['longitude_gps', 'latitude_gps']].apply(z_coord, axis=1)

train_dem['longlat'] = train_dem['longitude_gps'] + train_dem['latitude_gps']
test_dem['longlat'] = test_dem['longitude_gps'] + train_dem['latitude_gps']
```
